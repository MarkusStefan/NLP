```python
from keyphrase_vectorizers import KeyphraseCountVectorizer

docs = ["""Supervised learning is the machine learning task of learning a function that
         maps an input to an output based on example input-output pairs. It infers a
         function from labeled training data consisting of a set of training examples.
         In supervised learning, each example is a pair consisting of an input object
         (typically a vector) and a desired output value (also called the supervisory signal). 
         A supervised learning algorithm analyzes the training data and produces an inferred function, 
         which can be used for mapping new examples. An optimal scenario will allow for the 
         algorithm to correctly determine the class labels for unseen instances. This requires 
         the learning algorithm to generalize from the training data to unseen situations in a 
         'reasonable' way (see inductive bias).""", 
             
        """Keywords are defined as phrases that capture the main topics discussed in a document. 
        As they offer a brief yet precise summary of document content, they can be utilized for various applications. 
        In an information retrieval environment, they serve as an indication of document relevance for users, as the list 
        of keywords can quickly help to determine whether a given document is relevant to their interest. 
        As keywords reflect a document's main topics, they can be utilized to classify documents into groups 
        by measuring the overlap between the keywords assigned to them. Keywords are also used proactively 
        in information retrieval."""]
        
# Init default vectorizer.
vectorizer = KeyphraseCountVectorizer()

# Print parameters
print(vectorizer.get_params())
```

```python
# After initializing the vectorizer, it can be fitted
# to learn the keyphrases from the text documents.
vectorizer.fit(docs)
```

```python
# After learning the keyphrases, they can be returned.
keyphrases = vectorizer.get_feature_names_out()

print(keyphrases)
```

```python
import spacy, re
from keyphrase_vectorizers import KeyphraseCountVectorizer

# Load the German language model
nlp = spacy.load('de_core_news_lg')

# Sample text
# text = "Hans und Peter gingen in den Park. Anna und Maria spielten Fuball."

text = """

Absolut! Hier sind 5 Beispiele fr Beschwerde-E-Mails, die du an den Kundenservice deiner Bank richten knntest. Dabei habe ich verschiedene Szenarien bercksichtigt, um dir eine breite Palette an Mglichkeiten aufzuzeigen:

Beispiel 1: Fehlende berweisung

Sehr geehrte Damen und Herren,

mit diesem Schreiben mchte ich eine fehlende berweisung melden. Am [Datum] habe ich eine berweisung in Hhe von [Betrag] auf das Konto [Kontonummer] veranlasst. Bis heute wurde dieser Betrag jedoch noch nicht gutgeschrieben.

Meine berweisungsdaten lauten wie folgt:

Auftraggeber: [Dein Name]
Auftraggeber-Konto: [Deine Kontonummer]
Empfnger: [Name des Empfngers]
Empfnger-Konto: [Kontonummer des Empfngers]
Verwendungszweck: [Verwendungszweck]
Ich bitte Sie um dringende Klrung dieses Sachverhalts und um eine schnellstmgliche Gutschrift des fehlenden Betrags.

Mit freundlichen Gren,
[Dein Name]
[Deine Kundennummer]

Beispiel 2: Fehlerhafte Kontoauszge

Sehr geehrte Damen und Herren,

in meinen letzten Kontoauszgen habe ich mehrere Ungereimtheiten festgestellt. So wurden beispielsweise folgende Buchungen nicht korrekt erfasst:

[Beschreibung der ersten fehlerhaften Buchung]
[Beschreibung der zweiten fehlerhaften Buchung]
[usw.]
Ich bitte Sie, meine Kontoauszge zu berprfen und die Fehler umgehend zu korrigieren. Bitte senden Sie mir anschlieend korrigierte Kontoauszge zu.

Vielen Dank fr Ihre Bemhungen.

Mit freundlichen Gren,
[Dein Name]
[Deine Kundennummer]

Beispiel 3: Probleme mit der Online-Banking-Anwendung

Sehr geehrte Damen und Herren,

seit einigen Tagen habe ich Probleme bei der Nutzung meiner Online-Banking-Anwendung. Folgende Fehler treten auf:

[Beschreibung des ersten Problems]
[Beschreibung des zweiten Problems]
[usw.]
Ich habe bereits versucht, das Problem durch [beschriebene Lsungsversuche] zu beheben, jedoch ohne Erfolg.

Ich bitte Sie um Untersttzung bei der Lsung dieses Problems.

Mit freundlichen Gren,
[Dein Name]
[Deine Kundennummer]

Beispiel 4: Unberechtigte Abbuchung

Sehr geehrte Damen und Herren,

auf meinem Konto wurde am [Datum] eine Abbuchung in Hhe von [Betrag] vorgenommen, die ich nicht autorisiert habe. Der Zahlungsempfnger ist [Name des Zahlungsempfngers].

Ich bitte Sie, diese Abbuchung umgehend zu stornieren und den Betrag meinem Konto gutzuschreiben.

Bitte teilen Sie mir mit, welche weiteren Schritte ich unternehmen muss, um diesen Vorfall zu klren.

Mit freundlichen Gren,
[Dein Name]
[Deine Kundennummer]

Beispiel 5: Unzufriedenheit mit der Beratung

Sehr geehrte Damen und Herren,

am [Datum] habe ich mich in Ihrer Filiale in [Ort] bezglich [Thema] beraten lassen. Leider war ich mit der Beratung nicht zufrieden, da [Grnde fr die Unzufriedenheit].

Ich erwarte von Ihnen eine ausfhrlichere und verstndlichere Beratung zu diesem Thema.

Mit freundlichen Gren,
[Dein Name]
[Deine Kundennummer]

Wichtige Hinweise:

Sei konkret: Beschreibe dein Problem so genau wie mglich.
Sei hflich: Auch bei berechtigter Kritik solltest du hflich bleiben.
Sammle alle relevanten Unterlagen: Kontoauszge, berweisungsbelege, etc. knnen bei der Klrung helfen.
Behalte eine Kopie: Bewahre eine Kopie deiner E-Mail und aller weiteren Korrespondenz auf.
Zustzliche Tipps:

Nenne deine Kundennummer: So kann die Bank deine Anfrage schneller zuordnen.
Setze eine angemessene Frist: Gib der Bank eine realistische Frist fr die Bearbeitung deiner Beschwerde.
Nutze die Betreffzeile: Eine przise Betreffzeile hilft dabei, dass deine E-Mail schneller bearbeitet wird (z.B. "Fehlende berweisung am [Datum]").
Ich hoffe, diese Beispiele helfen dir dabei, eine klare und przise Beschwerde-E-Mail zu formulieren. Viel Erfolg!

"""

pattern = r'^\[\w\s\]+'
text = re.sub(pattern, '', text)


# Perform NER to identify people's names
doc = nlp(text)
names = [ent.text for ent in doc.ents if ent.label_ == 'PER']

# Remove people's names from the text
for name in names:
    text = text.replace(name, '')

# Initialize the vectorizer
vectorizer = KeyphraseCountVectorizer(
    spacy_pipeline='de_core_news_lg',
    pos_pattern='<ADJ.*>*<N.*>+',
    stop_words='german'
)

# Fit the vectorizer on the modified text
vectorizer.fit([text])

# Transform the text to get the keyphrases
# keyphrases = vectorizer.transform([text])
keyphrases = vectorizer.get_feature_names_out()

print(keyphrases)
```
